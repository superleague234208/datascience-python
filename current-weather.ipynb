```python
import requests
import os
import pandas as pd
import matplotlib.pyplot as plt

# This is fake key , You can get the actual one by creating an acocunt on api.openweathermap.org #
app_id="abcdefg1234567hijklmn8910" 
```

```python
city_name_list = []
city_maxtemp_list = []
city_mintemp_list = []
city_feelliketemp_list = []
city_windspeed_list = []
```

```python
def get_temp(city_name):
  
  url = "https://api.openweathermap.org/data/2.5/weather?q="+city_name+"&appid="+app_id+"&units=metric"
  response = requests.get(url)
  data=response.json()
  weather_list = []
  
  if data['cod'] == "404":
    weather_list = ['Invalid City']
    
  else:
    weather_list.append(data['main']['temp_max'])
    weather_list.append(data['main']['temp_min'])
    weather_list.append(data['main']['feels_like'])
    weather_list.append(data['wind']['speed'])
  
  return weather_list
```

```python
for i in range (1,6):
  city_name = input("Enter the city name "+str(i)+" : ")
  weather_list = get_temp(city_name)
  
  if(len(weather_list) == 4):
    city_name_list.append(city_name)
    city_maxtemp_list.append(weather_list[0])
    city_mintemp_list.append(weather_list[1])
    city_feelliketemp_list.append(weather_list[2])
    city_windspeed_list.append(weather_list[3])
  else:
    pass
```

```python
fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)

ax1.bar(city_name_list,city_maxtemp_list,color = "red",edgecolor = "black")
ax1.set(title="Max Temp (Celcius)",xlabel="City",ylabel="Temp(C)")
ax1.bar_label(ax1.containers[0])

ax2.bar(city_name_list,city_mintemp_list,color = "blue",edgecolor = "black")
ax2.set(title="Min Temp (Celcius)",xlabel="City",ylabel="Temp(C)")
ax2.bar_label(ax2.containers[0])

ax3.bar(city_name_list,city_feelliketemp_list,color = "green",edgecolor = "black")
ax3.set(title="Feels Like Temp (Celcius)",xlabel="City",ylabel="Temp(C)")
ax3.bar_label(ax3.containers[0])

ax4.bar(city_name_list,city_windspeed_list,color = "purple",edgecolor = "black")
ax4.set(title="Wind Speed (MPH)",xlabel="City",ylabel="Speed(MPH)")
ax4.bar_label(ax4.containers[0])



plt.tight_layout()
plt.show()
plt.savefig("Max_Temp.png")
```
